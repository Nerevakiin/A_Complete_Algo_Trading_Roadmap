```python
#Mean returns are misleading because 0.2 return would mean:
100*1.2*1.2
#Which gives 144.0 but we see that in reality it is 95
```

```python
logreturns.mean()
```

```python
100*np.exp(logreturns.mean()*2)
#This returns 95 which is correct. Logarithmic returns are much more accurate compared to simple returns
```

## Cummaltive Returns, Drawdowns etc...

```python
apple = yf.download("AAPL")
```

```python
apple = apple.Close
```

```python
apple
```

```python
#create a new column with the daily returns of AAPL
apple["d_returns"]=np.log(apple.div(apple.shift(1)))
```

```python
apple
```

```python
apple.dropna(inplace=True)
apple
```

```python
apple.d_returns.sum()
```

```python
np.exp(apple.d_returns.sum())
```

```python
apple["cummreturns"]=apple.d_returns.cumsum().apply(np.exp)
```

```python
apple
```

```python
apple.cummreturns.plot(figsize=(12,8), title="AAPL buy and hold", fontsize=12)
plt.show()
```

```python
apple.d_returns.mean()*252
```

```python
#standard deviation
apple.d_returns.std()*np.sqrt(252)
```

```python
# how to calculate drawdowns
apple["cummax"]=apple.cummreturns.cummax()
apple
```

```python
apple[["cummreturns", "cummax"]].plot(figsize=(12,8), title="AAPL buy and hold + cummax", fontsize=12)
plt.show()
```

```python
apple["drawdown"]=apple["cummax"]-apple["cummreturns"]
apple
```

```python
apple.drawdown.max()
```

```python
#find out when exactly did the maximum drawdown happen
apple.drawdown.idxmax()
```

```python
apple.loc[(apple.index<='2025-10-22')]
```

```python
apple["drawdown%"]=(apple["cummax"]-apple["cummreturns"])/apple["cummax"]
apple
```

```python
apple["drawdown%"].max()
```

```python
apple["drawdown%"].idxmax()
```

## SMA Strategy

```python
data = apple.Close.loc[(apple.index>='1991-01-01')]
data
```

```python
sma_s=50
sma_l=100
```

```python
data["sma_s"] = data.Close.rolling(sma_s).mean()
data["sma_l"] = data.Close.rolling(sma_l).mean()
```

```python
# First, check what columns you actually have
print(data.columns)

# # If you see MultiIndex like [('Close', 'AAPL')], then use:
# data["sma_s"] = data[('Close', 'AAPL')].rolling(sma_s).mean()
# data["sma_l"] = data[('Close', 'AAPL')].rolling(sma_l).mean()

# # Or flatten the columns to match the video:
# data.columns = ['Close']  # Rename to simple column name
# data["sma_s"] = data.Close.rolling(sma_s).mean()
# data["sma_l"] = data.Close.rolling(sma_l).mean()
```

```python
# Rename the column to 'Close' like in the video
data = data.rename(columns={'AAPL': 'Close'})

# Now this will work exactly like the video
data["sma_s"] = data.Close.rolling(sma_s).mean()
data["sma_l"] = data.Close.rolling(sma_l).mean()

print(data.head())
```

```python
data 
```

```python
data.plot(figsize=(12,6),title="APPLE - SMA{} | S<A{}".format(sma_s,sma_l), fontsize=12)
```

```python
data.loc["2016"].plot(figsize=(12,6),title="APPLE - SMA{} | SMA{}".format(sma_s,sma_l), fontsize=12)
```

```python
data.dropna(inplace=True)
data
```

```python
data["position"] = np.where(data["sma_s"] > data["sma_l"], 1, -1)
data
```

```python
data.loc["2016",["sma_s", "sma_l", "position"]].plot(figsize=(12,6),title="APPLE - SMA{} | SMA{}".format(sma_s,sma_l), fontsize=12, secondary_y = "position")
```

```python
data["returnsb&h"] = np.log(data.Close.div(data.Close.shift(1)))
data
```

```python
data["strategy"] = data["returnsb&h"] * data.position.shift(1)
data
```

```python
data.dropna(inplace=True)
```

```python
data[["returnsb&h","strategy"]].sum()
```

```python
data[["returnsb&h","strategy"]].sum().apply(np.exp) #what 1$ will be
```

```python
data[["returnsb&h","strategy"]].std()*np.sqrt(252) #annual std
```
