```python
aapl = close.AAPL.copy().to_frame()
```

```python
aapl
```

```python
aapl.shift(periods=1)
```

```python
#we do this to calculate the % change on daily price movement
aapl["lag1"] = aapl.shift(periods=1)
aapl
```

```python
#Numerical difference
aapl["Diff"] = aapl.AAPL.sub(aapl.lag1)
aapl
```

```python
#Percentage difference
aapl.AAPL.div(aapl.lag1)
```

```python
#adding the % change as a new column.
aapl["% Change"] = aapl.AAPL.div(aapl.lag1).sub(1).mul(100)
aapl
```

```python
aapl["Diff2"]=aapl.AAPL.diff(periods=1)
aapl
```

```python
aapl["% Change2"] = aapl.AAPL.pct_change(periods=1).mul(100)
aapl
```

```python
del aapl["% Change"]
aapl
```

```python
#rename the column from '%change2' to 'Change'
aapl.rename(columns = {'% Change2':'Change'}, inplace = True)
aapl
```

```python
#get last day of the Month
aapl.AAPL.resample("M").last()
```

```python
#last Business Day of the Month along with the % change on a monthly rate (.pct_change)
aapl.AAPL.resample("BM").last().pct_change(periods = 1).mul(100)
```

```python
del aapl["Change"]
```

```python
ret = aapl['AAPL'].pct_change().dropna()
ret.plot(kind="hist", figsize=(12,8), bins=100)
plt.show()
```

```python
#calculate mean return
daily_mean_return = ret.mean()
daily_mean_return
```

```python
var_daily = ret.var()
var_daily
```

```python
#standard deviation
std_daily = np.sqrt(var_daily)
std_daily
```

```python
annual_mean_ret = daily_mean_return*252 #working days
annual_mean_ret
```

```python
annual_var_return = var_daily*252
annual_var_return
```

```python
annual_std_returns= np.sqrt(annual_var_return)
annual_std_returns
```

```python
ticker=["SPY", "AAPL", "KO", "IBM", "DIS", "MSFT"]
```

```python
stocks = yf.download(ticker, start = "2010-01-01", end = "2021-01-01")
```

```python
stocks
```

```python
close = stocks.loc[:, "Close"].copy()
```

```python
normclose = close.div(close.iloc[0]).mul(100)
```

```python
normclose.plot(figsize=(10,8), fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

```python
ret= close.pct_change().dropna()
ret.head()
```

```python
#to ".T" antistrefei tis grammes me tis styles
ret.describe().T
```

```python
summary = ret.describe().T.loc[:,["mean", "std"]]
summary
```

```python
summary["mean"] = summary["mean"]*252
summary["std"] = summary["std"]*np.sqrt(252)
summary
```

```python
summary.plot.scatter(x="std", y="mean", figsize=(12,8), s = 50, fontsize = 15)
#a for loop which adds labels to the dots
for i in summary.index:
    plt.annotate(i, xy = (summary.loc[i, "std"]+0.002, summary.loc[i,"mean"]+0.002), size=15)    

#adds a label
plt.xlabel("Annual risk(std)", fontsize = 15)
plt.ylabel("Annual Return", fontsize = 15)
plt.title("Risk/return", fontsize = 25)
```
